This code is meant for convert a list of votes in a .txt file, divided by row as:

i-row: + voters
(i+1)-row: - votes
(i+2)-row: 0 votes

where i is a single resolution.

This code convert it in an adjacency matrix and save it in a .csv file.

```python
import numpy as np
import pandas as pd

meps = pd.read_excel('../data/meps.xlsx')
meps.politicalGroup = [x.replace('\n','') for x in meps.politicalGroup]
meps.head()
```

```python
# add coloumn
meps = meps.assign(idGroup = pd.Series(np.zeros((meps.shape[0])))) 
```

```python
for row in range(meps.shape[0]):
    if meps.at[row, 'politicalGroup'] == "Group of the European People's Party (Christian Democrats)":
        meps.at[row, 'idGroup'] = 1
    elif meps.at[row, 'politicalGroup'] == "Europe of Freedom and Direct Democracy Group":
        meps.at[row, 'idGroup'] = 2
    elif meps.at[row, 'politicalGroup'] == "Group of the Progressive Alliance of Socialists and Democrats in the European Parliament":
        meps.at[row, 'idGroup'] = 3
    elif meps.at[row, 'politicalGroup'] == "Group of the Alliance of Liberals and Democrats for Europe":
        meps.at[row, 'idGroup'] = 4
    elif meps.at[row, 'politicalGroup'] == "Europe of Nations and Freedom Group":
        meps.at[row, 'idGroup'] = 5
    elif meps.at[row, 'politicalGroup'] == "Confederal Group of the European United Left - Nordic Green Left":
        meps.at[row, 'idGroup'] = 6
    elif meps.at[row, 'politicalGroup'] == "Group of the Greens/European Free Alliance":
        meps.at[row, 'idGroup'] = 7
    elif meps.at[row, 'politicalGroup'] == "European Conservatives and Reformists Group":
        meps.at[row, 'idGroup'] = 8
    elif meps.at[row, 'politicalGroup'] == "Non-attached Members":
        meps.at[row, 'idGroup'] = 9
    else:
        meps.at[row, 'idGroup'] = 0     
```

```python
meps_list = [p.upper() for p in meps.fullName]  
votes = dict([(key, []) for key in meps_list])
votes
print(len(votes))
```

1: +<br>
-1: -<br>
0: neutral<br>
A: absent<br>

#### Adding votes

```python
f = open('./all-the-votes-ls-edit.txt', 'r', encoding="utf-8-sig") 
count = 0 # is the number of votes already processed
while(1): 
    
    ####### first: vote +
    y = f.readline()
    #print(y)
    if (y == ""): # I know it's horrible but it's work...
        break
    y = y[0:len(y)-1] # erase the \n last character
    y = y.split(', ')
    y = [p.upper() for p in y]
    for k in y:
        votes[k].append(1)

    ####### second: vote -
    n = f.readline()
    #print(n)
    n = n[0:len(n)-1] # erase the \n last character
    n = n.split(', ')
    n = [p.upper() for p in n]
    for k in n:
        if k in votes:
            votes[k].append(-1)

    
    ####### third: vote 0
    t = f.readline()
    #print(t)
    t = t[0:len(t)-1] # erase the \n last character
    t = t.split(', ')
    t = [p.upper() for p in t]
    for k in t:
        if k in votes:
            votes[k].append(0)

    
    ####### absent
    for k in votes:
        if len(votes[k]) < (count + 1):
            votes[k].append('A')

    len(votes)
    count = count + 1
    #print("done")

f.close()
```

```python
M = np.empty([DB.shape[0], DB.shape[0]+1], dtype=int)

#add id group in the last coloumn
meps.idGroup.astype(int) # as int coloumn

for row in np.arange(DB.shape[0]):
    M[row,854] = meps.at[row, 'idGroup']
```

```python
# (this cell is really slow in the computation)
for i in np.arange(DB.shape[0]):
    for j in np.arange(DB.shape[0]):
        #print(i,j)
        M[i,j] = sum(DB.iloc[i,:]==DB.iloc[j,:])-sum((DB.iloc[i,:]=='A')&(DB.iloc[j,:]=='A'))
```

```python
df = pd.DataFrame(M)
df.to_csv('../data/MwithGroup2.csv')
```
